* A base de dados utilizada é uma base pública do Portal Brasileiro de Dados Abertos disponível em https://dados.gov.br/dataset/resultados-2022 -> "PR - Votação por seção eleitoral - 2022";

* A base contém 3397518 linhas e 24 colunas

* Descrição das colunas:

  1. DT_GERACAO: data da geração do arquivo;
  2. HH_GERACAO: hora da geração do arquivo;
  3. ANO_ELEICAO: ano referência da eleição;
  4. CD_TIPO_ELEICAO: código do tipo de eleição;
  5. NM_TIPO_ELEICAO: nome do tipo de eleição;
  6. NR_TURNO: número do turno da eleição;
  7. CD_ELEICAO: código da eleição;
  8. DS_ELEICAO: descrição da eleição;
  9. DT_ELEICAO: data em que ocorreu a eleição;
  10. TP_ABRANGENCIA: tipo de abrangência da eleição;
  11. SG_UF: sigla da UF onde ocorreu a eleição;
  12. SG_UE: sigla da unidade eleitoral do candidato;
  13. NM_UE: nome da unidade eleitoral do candidato;
  14. CD_MUNICIPIO: código TSE do município onde ocorreu a eleição;
  15. NM_MUNICIPIO: nome do município onde ocorreu a eleição;
  16. NR_ZONA: número da zona eleitoral em que ocorreu a eleição;
  17. NR_SECAO: número da seção eleitoral em que ocorreu a eleição
  18. CD_CARGO: código do cargo de candidatura;
  19. DS_CARGO: descrição do cargo de candidatura;
  20. NR_VOTAVEL: número do candidato, número do partido, 95, 96, 97 ou 98;
  21. NM_VOTAVEL: nome do candidato, nome do partido, voto em branco, voto nulo ou voto anulado e apurado em separado
  22. QT_VOTOS: quantidade de voros referente ao tipo votável para o município e seção
  23. NR_LOCAL_VOTACAO: número do local de votação;
  24. SQ_CANDIDATO: sequência do candidato.

```python
# Cria um dicionário com o código do cargo e a descrição do cargo
cargos = votacao_2022[['CD_CARGO', 'DS_CARGO']].set_index('CD_CARGO').to_dict()
cargos = cargos['DS_CARGO']
cargos
```

```python
# Seleciona o código do cargo
cod_cargo_candidato = 3
cargo_candidato = cargos.get(cod_cargo_candidato)
cargo_candidato
```

```python
# Define um dataframe dos votos com os candidatos do cargo selecionado
votos_cargo = votacao_2022[votacao_2022['CD_CARGO'] == cod_cargo_candidato]
```

```python
# Cria um dicionário com o número e nome dos candidatos pelo cargo selecionado
candidatos = votos_cargo[['NR_VOTAVEL', 'NM_VOTAVEL']].set_index('NR_VOTAVEL').to_dict(orient='dict')
candidatos = candidatos['NM_VOTAVEL']

pd.DataFrame(list(candidatos.items()),columns = ['Número votável', 'Nome candidato'])
```

```python
# Seleciona o candidato
cod_candidato = 12
candidato = candidatos.get(cod_candidato)
candidato
```

```python
# Define um DF com os votos por Estados (esta base contém apenas votos do PR)
estado = 'PR'
votos_UF = votacao_2022[votacao_2022['SG_UF'] == estado]
```

```python
# Define um DF com os votos por município
municipio = 'CORBÉLIA'
votos_municipio = votos_UF[votos_UF['NM_MUNICIPIO'] == municipio]

print(f'{municipio} teve {len(votos_municipio)} votos')
```

```python
# Apresenta a quantidade de votos que o candidato conseguiu no município
votos_candidato_municipio = votos_municipio[votos_municipio['NR_VOTAVEL'] == cod_candidato]
print(f'O candidato "{candidato} - {cod_candidato}" conseguiu {len(votos_candidato_municipio)} em {municipio}')
```

```python
municipios = votos_UF['NM_MUNICIPIO'].unique()

print(f'Candidato: {candidato}. Número votável: {cod_candidato}')
for municipio in municipios:
  votacao_municipio = votacao_2022[votacao_2022['NM_MUNICIPIO'] == municipio]
  qtde_votos_candidato_municipio = votacao_municipio[votacao_municipio["NR_VOTAVEL"] == cod_candidato]
  print(f"Município: {municipio} = {len(qtde_votos_candidato_municipio)} votos")
```

```python
# Cidade com mais voto do candidato e cidade com menos voto do candidato
qtde_mais_votos = 0
cidade_mais_votos = ''
qtde_menos_votos = 0
cidade_menos_votos = []

print(f'Candidato: {candidato}. Número votável: {cod_candidato}')
print()
for municipio in municipios:
  votacao_municipio = votacao_2022[votacao_2022['NM_MUNICIPIO'] == municipio]
  qtde_votos_candidato_municipio = votacao_municipio[votacao_municipio["NR_VOTAVEL"] == cod_candidato]

  if qtde_mais_votos < len(qtde_votos_candidato_municipio):
    qtde_mais_votos = len(qtde_votos_candidato_municipio)
    cidade_mais_votos = municipio

  if len(qtde_votos_candidato_municipio) < qtde_menos_votos:
    qtde_menos_votos = len(qtde_votos_candidato_municipio)
    cidade_menos_votos.append(municipio)

print(f'A cidade com mais votos para {candidato} é {cidade_mais_votos} com {qtde_mais_votos} votos')
print(f'As cidades com menos votos para {candidato} são: {len(cidade_menos_votos)} cidades:')
for cidade in cidade_menos_votos:
  print(f'{cidade} com {qtde_menos_votos} voto(s)')
```
